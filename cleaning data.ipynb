```python
df.date_added
```

```python
def date_convet(date):
    dates=pd.to_datetime(date)
    return dates.strftime("%d:%m:%y")
df.date_added=df.date_added.apply(date_convet)
```

# conveted the date format to date:month:year using the strftime fuction
